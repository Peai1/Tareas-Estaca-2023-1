### 1)

Para poder determinar el número posible de caminos distintos que tiene Bowser para llegar al destino (m,n) se puede dar con la siguiente formula:


<p style="text-align:center; font-size:40px">
$\binom{m+n}{n}$
</p>


Esto se debe a que si denominamos m las veces que se debe mover hacia derecha y n las veces que se debe de mover hacia arriba, en total para poder llegar al destino debe moverse m+n veces, de aqui necesitamos elegir n movimientos hacia arriba, por lo que los caminos posibles se puede obtener de la cantidad de formas de elegir n movimientos entre m+n pasos. Todas las combinaciones de caminos posibles están dadas por la siguiente combinatoria. 


Si m = 100 y n = 100, entonces usando la combinatoria, se obtienen la siguiente cantidad de caminos

```python
import math
m = 100
n = 100

combinatoria = math.comb(m + n, n)
print(combinatoria)
```

### 2)

Para determinar la cantidad de caminos distintos en los que bowser puede llegar al territorio de mario, pasando sí o sí por los puntos mencionados (checkpoints), se realiza un analisis similar al de la pregunta anterior, sin embargo se calculan la cantidad de caminos desde el punto inicial hasta el primer checkpoint, luego la cantidad de caminos desde el primer checkpoint al segundo, y así sucesivamente. Finalmente se multiplican estos resultados dado que se pueden realizar combinaciones de los caminos entre checkpoints. Como indica la sgte. fórmula:

<p style="text-align:center; font-size:40px">
$caminos=\binom{checkX_0+checkY_0}{checkY_0}\> * \> \prod_{i=1}^n\binom{(checkX_i-checkX_{i-1})\>+\> (checkY_i-checkY_{i-1})}{checkY_i-checkY_{i-1}}$

Donde $checkX_i$ es la cordenada X y $checkY_i$ la coordenada Y del checkpoint i, además por simplicidad se agrega m y n a la lista de los checkpoints, siendo este el checkpoint numero n en la productoria.  
</p>
Con m = 100 y n = 100 y los checkpoints del enunciado, al utilizar la fórmula descrita se obtienen la sgte cantidad de caminos: 

```python
from math import comb

m = 100
n = 100

checkpoints = [(11, 13), (29, 31), (53, 47), (71, 67), (89, 97),(m,n)]

suma_total = comb(checkpoints[0][0] + checkpoints[0][1], checkpoints[0][0])

for i in range(1, len(checkpoints)):
    diff_m = checkpoints[i][0] - checkpoints[i-1][0]
    diff_n = checkpoints[i][1] - checkpoints[i-1][1]
    diff_sum = diff_m + diff_n
    diff_nn = checkpoints[i][0] - checkpoints[i-1][0]
    resul = comb(diff_sum, diff_nn)
    suma_total *= resul

print(suma_total)
```

### 3)

La función de densidad de probabilidad (FDP) está definida de la siguiente manera:

Sea X: la cantidad de pasos que da bowser para llegar al territorio de mario

El recorrido de la variable empieza de {m + n, m + n + 1, ...}

$$
\text{fdp}(x) = \begin{cases}
    d^m \cdot a^n \cdot \binom{x}{n} & \text{si } x = m + n \\
    d^{x-n} \cdot a^n \cdot \binom{x-1}{n-1} + d^m \cdot a^{x-m} \cdot \binom{x-1}{m-1} & \text{si } x > m + n \\
    0 & \text{en otro caso}
\end{cases}
$$

donde d es la probabilidad de moverse hacia la derecha y a es la probabilidad de moverse hacia arriba.

Cuando x = m + n, existe una unica forma de realizar esa cantidad de pasos, llegando al punto exacto (m,n), donde debe dar m pasos hacia la dereha y n pasos hacia arriba y para ver todas las posibles combinaciones para poder llegar a ese punto, se utiliza la formula obtenida en la pregunta 1) que es representada por la combinatoria

Nos damos cuenta de que hay dos formas para realizar una cantidad x de pasos (con x > m+n), ya sea por la izquierda o por abajo del territorio de mario.

En la fdp, cuando x > m + n, la primera parte representa la probabilidad de realizar x pasos llegando por abajo del territorio de mario y la segunda parte llegando por la izquierda del territorio de mario. Si tomamos como ejemplo m,n = 100,100 y x = 201, primero vemos todas las combinaciones posibles para llegar al punto (201,199), que corresponde a la combinatoria $\binom{x-1}{n-1}$ en donde ha dado 200 pasos y eso es multiplicado por la probabilidad correspondiente a los pasos a la derecha y hacia arriba que debe dar, en este caso con x = 201, debe dar los n pasos hacia arriba y x-n pasos hacia la derecha. En el caso de que llegue con 201 pasos por el territorio a la izquierda de mario, vemos las combinacions posibles para llegar al punto (199,201), dada por la combinatoria $\binom{x-1}{m-1}$, donde ha dado 200 pasos y eso multiplicarlo por las cantidades de pasos hacia la derecha y arriba que debe dar. La suma de esas probabilidades da la probabilidad total de llegar con 201 pasos en este caso pero funciona para cualquier cantidad de pasos x.

<img src='https://www.inf.utfsm.cl/~jnancu/Mario.jpg' width='360' style='display:block; margin:auto;'/>


```python
m = 100
n = 100

#recorrido de x = m+n, m+n+1, m+n+2, ...

a = 0.5
d = 0.5

def fdp(x):
  if x == (m+n):
    resul = ((a)**(m) * (d)**(n) * comb(x,n))
    return resul
  elif x > (m+n):  #der     arriba                            der          arriba
    resul =  ((d)**(x-n) * (a)**(n) * comb(x-1,n-1)) + ((d)**(m) * (a)**(x-m) * comb(x-1,m-1)) # primera parte es por abajo territorio mario, segunda parte por izq territorio mario
    return resul
  else:
    return 0

limite = 10000
sum = 0

for i in range(m+n,limite):
  sum += i * fdp(i)

print("Valor esperado de pasos :",sum)

# Crear los valores de x en el rango deseado
eje_x = np.arange(m+n, m+n+82)

# Calcular los valores correspondientes de la fdp para cada x
eje_y = [fdp(x) for x in eje_x]

# Graficar la fdp
plt.figure()
df_ej2_teo = pd.DataFrame({'x': eje_x, 'f(x) teórica': eje_y})
df_ej2_teo.head(10)
df_ej2_teo.plot(x="x", y= 'f(x) teórica', kind="bar", rot = 45, figsize=(26, 6), legend = False, ylabel = "f(x) teórica")
plt.show()
```

### 4)

Simulación de la fdp

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def simulacion_pasos_bowser(m, n, numero_simulaciones):
    pasos_graf = []
    total_pasos = 0
    resultados_ej2 = {'x': [], 'f(x) experimental': []}

    for _ in range(numero_simulaciones):
        x = 0
        y = 0
        pasos = 0

        while x < m or y < n:
            direccion = np.random.choice(["right", "up"],p=[0.5,0.5])
            if direccion == "right":
                x += 1
            elif direccion == "up":
                y += 1
            pasos += 1

        pasos_graf.append(pasos)
        total_pasos += pasos
    
    for elem in sorted(set(pasos_graf)):
      frecuencia = pasos_graf.count(elem)
      frecuencia_relativa = frecuencia / numero_simulaciones
      
      resultados_ej2['x'].append(elem)
      resultados_ej2['f(x) experimental'].append(frecuencia_relativa)
    
    # Graficar fdp
    df_ej2_sim = pd.DataFrame(resultados_ej2)
    plt.figure()
    df_ej2_sim.head(10)
    df_ej2_sim.plot(x="x", y= 'f(x) experimental', kind="bar", rot = 0, figsize=(30, 6), ylabel = "f(x) simulación")
    plt.show()
    # Pasos estimados
    pasos_promedio = total_pasos / numero_simulaciones
    return pasos_promedio, resultados_ej2

numero_simulaciones = [1000, 10000, 100000, 1000000]
#numero_simulaciones = [1000]
m = 100
n = 100

for simulaciones in numero_simulaciones:
    pasos_estimados, resultados_ej2 = simulacion_pasos_bowser(m, n, simulaciones)
    print(f"Simulaciones: {simulaciones}\nPasos estimados: {pasos_estimados}\n")
```

De aqui podemos observar que a medida que el numero de simulaciones aumenta, la fdp experimental se asemeja mucho más a la fdp teórica obtenida en la pregunta anterior, esto debido a la gran cantidad de simulaciones que aproxima de mejor manera lo que sucede realmente

### 5)

Gráfico interactivo y valor esperado en función de la probabilidad de ir hacia arriba

```python
from ipywidgets import interact, FloatSlider
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import comb

m = 100
n = 100
limite = 10000

def fdp(x,a):
  d = 1 - a
  if x == (m+n):
    resul = ((a)**(m) * (d)**(n) * comb(x,n))
    return resul
  elif x > (m+n):  #der     arriba                            der          arriba
    resul =  ((d)**(x-n) * (a)**(n) * comb(x-1,n-1)) + ((d)**(m) * (a)**(x-m) * comb(x-1,m-1)) # primera parte es por abajo territorio mario, segunda parte por izq territorio mario
    return resul
  else:
    return 0

def valor_esperado(a):
    sum = 0
    for i in range(m+n,limite):
      sum += i * fdp(i,a)
    return sum

def grafico_valor_esperado():
    valores_a = np.arange(0.05, 1.0, 0.05)
    # Calcula el valor esperado para cada valor de a
    plt.figure()
    valores_esperados = [valor_esperado(x) for x in valores_a]
    plt.plot(valores_a, valores_esperados)
    plt.xlabel('Probabilidad de ir hacia la arriba (a)')
    plt.ylabel('Valor Esperado')
    plt.title('Valor Esperado en función de la probabilidad de ir hacia la arriba')
    plt.grid(True)
    plt.show()

def grafico_fdp(a):
    
    if 0.25 < a < 0.75:
      lim = m+n+500
    else:
      lim = m+n+2700
    eje_x = np.arange((m+n)-50, lim)

    # Calcular los valores correspondientes de la fdp para cada x
    eje_y = [fdp(x,a) for x in eje_x]

    # Graficar la fdp
    plt.figure()
    df_ej2_teo = pd.DataFrame({'x': eje_x, 'f(x) teórica': eje_y})
    df_ej2_teo.head(10)
    df_ej2_teo.plot(x="x", y= 'f(x) teórica', kind="line", rot = 0, figsize=(18, 6), legend = False, ylabel = "f(x) teórica")
    plt.show()

interact(grafico_fdp, a=FloatSlider(value=0.5, min=0.05, max=0.97, step=0.05))
grafico_valor_esperado()
```

De aqui podemos realizar la siguiente conclusión:
Como se puede observar, cuando la probabilidad de ir hacia arriba varía ya sea aumentando o disminuyendo, se tiene un comportamiento similar. 
Cuando la probabilidad de ir hacia arriba es muy baja, la cantidad de pasos estimados aumenta drasticamente, esto debido a que es muy poco probable que Bowser haga un movimiento hacia arriba y la gran mayoria de movimientos serán hacia la derecha y en este caso Bowser tarda mucho pasos más en llegar al territorio de Mario, lo cual es más probable que llegue por abajo del territorio.
En cambio, si se aumenta la probabilidad de ir hacia arriba, ocurre algo similar ya que la gran mayoria de pasos que dará Bowser serán hacia arriba y es muy poco probable dar los m pasos hacia la derecha necesarios como minimo para llegar al territorio, en este caso sería más probable que Bowser llegue por la izquierda del territorio de Mario.
Del gráfico del valor esperado se puede observar una especie de "espejo" a partir de la probabilidad de ir hacia arriba = 0.5, si la probabilidad varía la misma cantidad disminuyendo o aumentando a partir del 0.5, el valor esperado de los pasos posee el mismo valor.

Para obtener esta probabilidad lo haremos en función de la cantidad de pasos que da Bowser para llegar al territorio de Mario.

Primero observamos que dada una cantidad de pasos "n" que se dan para llegar al territorio de bowser, la cantidad que veces que pueden ocurrir 4 movimientos consecutivos que alejan a Bowser serán n - 3.

La probabilidad de que Bowser se mueva en la dirección contraria a su objetivo en un solo paso es la suma de las probabilidades de moverse hacia la izquierda y hacia abajo, que es 0.3 + 0.1 = 0.4.

El evento de que Bowser decida abandonar su viaje es que se produzcan 4 'fracasos' (movimientos en dirección contraria) consecutivos. La probabilidad de 4 fracasos consecutivos es (0.4)^4 = 0.0256. Esto se puede obtener de la suma de las 16 combinaciones diferentes que hay para dar pasos a la izquierda y pasos hacia abajo.

Ahora, si denominamos P como la probabilidad de dar 4 fracasos consecutivos en n pasos y por ende abandonar, tenemos los siguientes casos base \
P(0) = P(1) = P(2) = P(3) = 0 \
P(4) = 0.0256

La probabilidad para cualquier cantidad de pasos que da es la siguiente.
Primero, cuando $n < 9$

$P(n) = \sum_{i=4}^{n-1} [P(i)] + 0.6 \times 0.0256$

El 0.6 viene ya que antes de los 4 pasos consecutivos tiene que haber un paso hacia a la derecha o hacia arriba, el 0.0256 de los 4 pasos consecutivos que lo alejan y la sumatoria corresponde a todos los posibles subconjuntos de 4 movimientos que pueden ser que lo alejen a bowser del territorio de mario.

Cuando $n \geq 9$, debemos ver el caso de que al principio no se hayan dado 4 movimientos que alejan a bowser consecutivos lo que está dado por 1 - 0.0256, por lo que la probabilidad cuando $n \geq 9$ es:

$P(n) = \sum_{i=4}^{n-1} [P(i)] + 0.6 \times 0.0256 \times (1-0.0256)^{n-8}$

Para obtener la probabilidad que queremos debemos multiplicarlo por la probabilidad de dar n pasos, que se obtuvo mediante la fdp de la pregunta 6), por lo que la probabilidad queda

$P(n) =
\begin{cases}
    \left(\sum_{i=4}^{n-1} [P(i)] + 0.6 \times 0.0256\right) \times f(n), & \text{si } n < 9 \\
    \left(\sum_{i=4}^{n-1} [P(i)] + 0.6 \times 0.0256 \times (1-0.0256)^{n-8}\right) \times f(n), & \text{si } n \geq 9 \\
\end{cases}
$

donde f representa la fdp de la pregunta 6


### 10)

Simulación de la probabilidad de que Bowser abandone.

Bowser abandona al dar 4 movimientos consecutivos que lo alejen del territorio de mario

```python
import random

def simulacion_bowser(m, n):
    x, y = 0, 0  # Coordenadas iniciales de Bowser

    alejamientos_consecutivos = 0  # Contador de movimientos fallidos consecutivos
    while True:
        # Generar un movimiento aleatorio
        direccion = random.choices(["up", "down", "right", "left"], weights=[0.4, 0.3, 0.2, 0.1])[0]
        # Actualizar las coordenadas según el movimiento
        if direccion == "up":
            y += 1
        elif direccion == "down":
            y -= 1
        elif direccion == "right":
            x += 1
        elif direccion == "left":
            x -= 1

        # Verificar si Bowser llegó al territorio de Mario
        if x >= m and y >= n:
            return False  # Bowser llegó al territorio de Mario

        # Verificar si Bowser debe abandonar el viaje
        if direccion in ["down", "left"]:
            alejamientos_consecutivos += 1
            if alejamientos_consecutivos == 4:
                return True  # Bowser abandona el viaje

        # Reiniciar el contador de movimientos fallidos consecutivos si Bowser se mueve hacia arriba o hacia la derecha
        else:
            alejamientos_consecutivos = 0


# Definir las coordenadas del territorio de Mario
m, n = 100, 100
# con 10 y 10 pasos estimados 139
# con 30 y 10 pasos estimados 307
# Simular el movimiento de Bowser para obtener una muestra de las probabilidades de abandono
numero_simulaciones = 100000
contador_abandono = 0

for _ in range(numero_simulaciones):
    if simulacion_bowser(m, n):
        contador_abandono += 1

# Calcular la probabilidad de abandono
probability = contador_abandono / numero_simulaciones

# Imprimir la probabilidad de abandono
print(f"Probabilidad de abandono: {probability}")
```
